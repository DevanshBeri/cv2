<h1> Creating Bind Trackbar </h1
>

```python
# creating a black image and window

def nothing(value):  # . This argument represents the current value of the trackbar
    print("Trackbar value:", value)  # will show in terminal


img = np.zeros((300, 512, 3), np.uint8)
cv2.namedWindow("image")

# cv2.createTrackbar(trackbarName, windowName, initialValue, maxValue, onChange)

cv2.createTrackbar("B", 'image', 0, 255, nothing)

cv2.createTrackbar("G", 'image', 0, 255, nothing)

cv2.createTrackbar("R", 'image', 0, 255, nothing)

# adding switch

cv2.createTrackbar('switch', 'image', 0, 1, nothing)


while (1):
    cv2.imshow("image", img)

    # if cv2.waitkey(1)==ord('q') window will close when press q

    if cv2.waitKey(1) >= 0:  # will close the window using any key
        break
    # to modify the black image or making trackbar work
    b = cv2.getTrackbarPos("B", "image")
    g = cv2.getTrackbarPos("G", "image")
    r = cv2.getTrackbarPos("R", "image")
    s = cv2.getTrackbarPos('switch', "image")

    if s == 0:
        img[:] = 0  # color will not change
    else:
        img[:] = [b, g, r]  # color will change
cv2.destroyAllWindows()
```

<h1> Creating Bind Trackbar  for image</h1> 

```python
import cv2


def nothing(value):
    print("Trackbar value:", value)


cv2.namedWindow("image")

cv2.createTrackbar("CP", 'image', 0, 500, nothing)

cv2.createTrackbar("B", 'image', 0, 255, nothing)

cv2.createTrackbar("G", 'image', 0, 255, nothing)

cv2.createTrackbar("R", 'image', 0, 255, nothing)


# adding switch
cv2.createTrackbar('switch', 'image', 0, 1, nothing)

while True:
    img = cv2.imread(
        r"C:\Users\DEVANSH\Desktop\CV2\images\Screenshot (15).png")

    image = cv2.resize(img, (1280, 720))

    pos = cv2.getTrackbarPos("CP", 'image')

    font = cv2.FONT_HERSHEY_SIMPLEX

    b = cv2.getTrackbarPos("B", "image")
    g = cv2.getTrackbarPos("G", "image")
    r = cv2.getTrackbarPos("R", "image")

    cv2.putText(image, str(pos), (640, 360), font, 4, (b, g, r), thickness=4,
                lineType=cv2.LINE_AA)

    cv2.imshow("image", image)

    if cv2.waitKey(1) >= 0:
        break

    s = cv2.getTrackbarPos("switch", "image")
    if s == 0:
        img[:] = [b, g, r]  # color will change
    else:
        img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

cv2.destroyAllWindows()
```

<h1>cv2.findContours() MOTION DETECTION</h1>

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread(r"C:\Users\DEVANSH\Desktop\CV2\images\Screenshot (15).png")

# Resize the image
image = cv2.resize(image, (1280, 720))

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert grayscale image to binary image
ret, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

# Find contours
contours, hierarchy = cv2.findContours(
    binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Create a copy of the image
image_copy = image.copy()

# Draw contours on the copy of the image
cv2.drawContours(image_copy, contours, -1, (0, 255, 0),
                 thickness=3, lineType=cv2.LINE_4)

# Display the images
cv2.imshow("Grayscale image", gray)
cv2.imshow("Draw Contours", image_copy)
cv2.imshow("Binary image", binary)

# Wait for a key press and destroy windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the image
image = cv2.imread(r"C:\Users\DEVANSH\Desktop\CV2\images\circle.png")

# Resize the image
image = cv2.resize(image, (600, 600))

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Convert grayscale image to binary image using Otsu's thresholding
ret, binary = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Find contours in the binary image
contours, hierarchy = cv2.findContours(
    binary, mode=cv2.RETR_LIST, method=cv2.CHAIN_APPROX_SIMPLE)

# Create a copy of the original image
image_copy = image.copy()

# Draw contours on the copy of the image
image_copy = cv2.drawContours(
    image_copy, contours, -1, (0, 255, 0), thickness=3, lineType=cv2.LINE_4)

# Display the grayscale image
cv2.imshow("Grayscale image", gray)

# Display the image with contours
cv2.imshow("Draw Contours", image_copy)

# Display the binary image
cv2.imshow("Binary image", binary)

# Wait for a key press to exit
cv2.waitKey(0)

# Close all open windows
cv2.destroyAllWindows()
```

<h1>MOTION DETECTION </h1>

```python
import cv2
import numpy as np

video = cv2.VideoCapture(r"C:\Users\DEVANSH\Desktop\CV2\video\people.mp4")

ret, frame1 = video.read()
ret, frame2 = video.read()

while ret:
    diff = cv2.absdiff(frame1, frame2)
    gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
    ret, binary = cv2.threshold(
        gray, 0, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

    contours, hierarchy = cv2.findContours(
        binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        (x, y, w, h) = cv2.boundingRect(contour)
        if cv2.contourArea(contour) < 700:
            continue
        cv2.rectangle(frame1, (x, y), (x+w, y+h), (0, 255, 0), thickness=5)
        cv2.putText(frame1, 'status: {}'.format('Movement'), (10, 20),
                    cv2.FONT_HERSHEY_COMPLEX, fontScale=1, color=(0, 255, 0), thickness=2)

    image = cv2.resize(frame1, (1280, 720))
    cv2.imshow("image", image)
    frame1 = frame2

    ret, frame2 = video.read()
    if cv2.waitKey(1) >= 0:
        break

cv2.destroyAllWindows()
```
